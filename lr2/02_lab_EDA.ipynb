8. Построение графиков зависимости 5 любых переменных от целевой переменной.

```python
target_variable = "G1"
other_variables = ["G2", "G3"]

for var in other_variables:
    plt.figure(figsize=(8, 6))
    plt.scatter(data[var], data[target_variable], alpha=0.5)
    plt.title(f"Зависимость {target_variable} от {var}")
    plt.xlabel(var)
    plt.ylabel(target_variable)
    plt.grid(True)
    plt.show()
```

```python
target_variable = "G1"
other_variables = ["Medu", "Fedu"]

for var in other_variables:
    plt.figure(figsize=(8, 6))
    data.boxplot(column=target_variable, by=var)
    plt.title(f"Зависимость {target_variable} от {var}")
    plt.xlabel(var)
    plt.ylabel(target_variable)
    plt.grid(True)
    plt.show()
```

```python
target_variable = "G1"
other_variables = ["sex"]

for var in other_variables:
    plt.figure(figsize=(8, 6))
    data.boxplot(column=target_variable, by=var)
    plt.title(f"Зависимость {target_variable} от {var}")
    plt.xlabel(var)
    plt.ylabel(target_variable)
    plt.grid(True)
    plt.show()
```
